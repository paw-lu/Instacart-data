```python
import pandas as pd
import re
```

```python
products = pd.read_csv('../instacart_2017_05_01/products.csv')
departments = pd.read_csv('../instacart_2017_05_01/departments.csv')
aisles = pd.read_csv('../instacart_2017_05_01/aisles.csv')
order_products_prior = pd.read_csv('../instacart_2017_05_01/order_products__prior.csv')
orders = pd.read_csv('../instacart_2017_05_01/orders.csv')
product_info = products.merge(departments, on='department_id', how='inner')
product_info = product_info.merge(aisles, on='aisle_id', how='inner')
product_info = product_info[['product_id', 'product_name', 'department', 'aisle']]
product_info.head()
```

```python
orders = orders[orders['eval_set'] == 'prior']
orders.head()
```

```python
def product_search(food_regex, df):
    f = open('output.txt', 'w')
    f.write(food_regex + "\n")
    def food_finder(product):
        if re.search(food_regex, product['product_name']) is not None:
            f.write(str(product['product_id']) + " ")
            print(product['product_id'], product['product_name'])
        
    df.apply(food_finder, axis=1)
    f.close()
```

```python
product_orders = orders.merge(order_products_prior, on='order_id', how='inner')
```

```python
product_orders['product_id'].value_counts()
```

```python
popular_items = product_orders['product_id'].value_counts()
```

```python
popular_items = popular_items.iloc[:500].index.tolist()
```

```python
popular_items
```

```python
f = open('popular_items.txt', 'w')
f.write('[')
for i in popular_items:
    f.write(str(i) + ", ")
f.write(']')
f.close()
```
